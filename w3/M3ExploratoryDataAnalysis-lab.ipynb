## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(df['ConvertedComp'])
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
df['ConvertedComp'].describe()
```

Find out the upper and lower bounds.


```python
import numpy as np
import pandas as pd 
import scipy.stats as stats

# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = df['ConvertedComp'].apply(stats.iqr)
#print('{}, {}'.format(Q1, Q3))
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
#only keep rows in dataframe that have values within 1.5*IQR of Q1 and Q3
# 
sum((df['ConvertedComp'] < (Q1 - 1.5 * IQR)) |(df['ConvertedComp'] > (Q3 + 1.5 * IQR)))

# df = df[~((df['ConvertedComp'] < (Q1 - 1.5 * IQR)) |(df['ConvertedComp'] > (Q3 + 1.5 * IQR))).any()]
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.

```python
# your code goes here
df_cleaned = df[~((df['ConvertedComp'] < (Q1-1.5*IQR)) | (df['ConvertedComp'] > (Q3+1.5*IQR)))]
sns.boxplot(df_cleaned['ConvertedComp'])
```

```python
df_cleaned['ConvertedComp'].median()
```
